```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('a2_data/housepricedata.csv')
df
```

# To normalize the data

                     zi = (xi−min(x)) / (max(x)−min(x))

where x=(x1,...,xn) and zi is now your ith normalized data. This will scale down the data to range 0 to 1.

```python
# To implement min-max normalization
normalized_df = (df-df.min()) / (df.max()-df.min())
print(normalized_df)

dataset_array = normalized_df.values
dataset_array
```

```python
# To split the array of 11 columns into 10 and 1 column
# split features and the target variable
X = dataset_array[:,0:10]
Y = dataset_array[:,10]

Y = Y.astype(int)
Y
```

```python
#implementing train test split
train_split_percent = 80
test_split_percent = 20

train_X , test_X = X[ : int( (train_split_percent/100)*X.shape[0]), :] , X[int( (train_split_percent/100)*X.shape[0]) : , : ]
train_Y , test_Y = Y[ : int( (train_split_percent/100)*X.shape[0])] , Y[int( (train_split_percent/100)*X.shape[0]) : ]
print(train_X.shape)
print(test_X.shape)
```
